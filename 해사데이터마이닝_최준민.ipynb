```python
# 1. 라이브러리 및 데이터 불러오기
import pandas as pd

train = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part2/ch6/diabetes_train.csv")
test = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part2/ch6/diabetes_test.csv")

# 2. 탐색적 데이터 분석(EDA)
#데이터 크기 확인
print("===== 데이터 크기 =====")
print("Train Shape:", train.shape)
print("Test Shape:", test.shape)
print("\n")

#컬럼별 자료형 확인
print("===== 데이터 정보(자료형) =====")
print(train.info())
print("\n")

#결측치 확인
print("===== train 결측치 수 =====")
print(train.isnull().sum().sum())
print("\n")

#target 확인 -> 분류 group 개수 확인
print("===== target 빈도 =====")
print(train['Outcome'].value_counts())

# 3. 데이터 전처리
target = train.pop('Outcome')

# 4. 검증 데이터 분할
from sklearn.model_selection import train_test_split
X_tr, X_val, y_tr, y_val = train_test_split(train, target, test_size=0.2, random_state=0)

print("\n ===== 분할된 데이터 크기 =====")
print(X_tr.shape, X_val.shape, y_tr.shape, y_val.shape)

#5. 머신러닝 학습 및 평가
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state=0)
rf.fit(X_tr, y_tr)
pred = rf.predict_proba(X_val)

#6. 예측 및 결과 파일 생성
pred = rf.predict_proba(test)
submit = pd.DataFrame({'pred': pred[:,1]})
submit.to_csv("result.csv", index=False)

# 제출파일 확인
print("\n ===== 제출파일 (샘플 5개) =====")
print(pd.read_csv("result.csv").head())
```

탐색적 데이터 분석(EDA)은 데이터 분석 문제 해결 과정에서 매우 중요한 단계이며, 특히 빅데이터 분석기사 실기시험과 같은 평가에서는 문제 해결 능력의 핵심적인 부분입니다. EDA가 필요한 합리적인 이유는 다음과 같습니다:

데이터 이해 및 파악: 데이터셋의 구조(컬럼 수, 데이터 타입 등), 내용, 분포 및 특성을 전반적으로 이해하고 파악하는 데 도움을 줍니다. 이를 통해 어떤 데이터가 있는지, 각 컬럼이 무엇을 의미하는지 알 수 있습니다.

문제점 및 이상치 발견: 데이터 내에 존재하는 결측치, 이상치(Outlier), 오류 또는 불일치하는 값을 식별할 수 있습니다. 이러한 문제점들을 미리 파악하고 처리하지 않으면 모델의 성능에 부정적인 영향을 미칠 수 있습니다.

패턴 및 관계 탐색: 데이터 간의 숨겨진 패턴, 추세, 상관관계 또는 그룹핑을 발견할 수 있습니다. 이는 데이터를 더 깊이 이해하고, 가설을 수립하며, 이후 모델링 방향을 설정하는 데 중요한 기반이 됩니다.

피처 엔지니어링 및 변수 선택 가이드: EDA를 통해 얻은 통찰력은 어떤 변수를 사용하고, 어떤 새로운 변수를 생성해야 할지(피처 엔지니어링), 또는 어떤 변수를 제외해야 할지 결정하는 데 중요한 단서를 제공합니다.

적절한 모델 선택에 기여: 데이터의 특성(선형성, 분포 등)을 이해하면 문제 해결에 가장 적합한 머신러닝 모델을 선택하는 데 도움이 됩니다.

요약하자면, EDA는 데이터를 '알아가는 과정'으로, 이는 단순히 데이터를 시각화하는 것을 넘어, 데이터의 본질을 이해하고, 예측 모델의 성능을 향상시키기 위한 필수적인 전처리 및 전략 수립 과정의 초석이 됩니다.
